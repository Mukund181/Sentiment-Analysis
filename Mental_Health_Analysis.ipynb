# Mental Health Sentiment Analysis (Rule-Based BERT)
This notebook uses a local `DistilBERT` model to detect emotions and responds with pre-defined empathetic mental-health coping strategies

```python
import pandas as pd
import random
from transformers import pipeline

# 1. Load the dataset (for context & testing)
try:
    df = pd.read_csv('dataset.txt')
    print("✅ Dataset Loaded. Sample:")
    display(df.head())
except Exception as e:
    print("Error loading dataset:", e)

# 2. Load the BERT Emotion Classifier
print("\nLoading BERT Emotion Classifier (This may take a few seconds)...")
emotion_classifier = pipeline("text-classification", model="bhadresh-savani/distilbert-base-uncased-emotion")
print("✅ BERT Model loaded successfully!")

# 3. Rule-Based Empathy Dictionary
empathy_responses = {
    "sadness": [
        "I'm really sorry you're feeling this way. Remember it's okay not to be okay. Trying a 5-minute distraction like listening to a comfort song might help ease the weight.",
        "That sounds incredibly heavy to carry. Please know you are not alone in this. Have you tried doing one small, comforting activity for yourself today?",
        "I hear you, and it's completely valid to feel sad. Giving yourself grace and taking things one hour at a time is sometimes the best strategy."
    ],
    "joy": [
        "That's wonderful to hear! Hold onto this positive feeling, it's so important to recognize our good days.",
        "I'm so glad things are going well for you right now! Let's celebrate these little victories.",
        "It sounds like you're in a great state of mind! Enjoy this moment."
    ],
    "love": [
        "It sounds like you're feeling very connected and affectionate right now, which is beautiful.",
        "Love is such a powerful emotion. Experiencing it fully is wonderful for your mental wellbeing.",
        "That warmth you're describing is a lovely thing. Keep leaning into those positive connections."
    ],
    "anger": [
        "It's completely okay to feel frustrated and angry. Let's take a deep breath together. A short walk might help clear some of that intense energy.",
        "I can understand why you'd feel that way. Experiencing anger is human. Trying some deep, slow breathing or writing your feelings down can help release the pressure.",
        "That sounds very frustrating. When you feel overwhelmed with anger, stepping away from the situation for 5 minutes of quiet can be a powerful coping tool."
    ],
    "fear": [
        "Feeling anxious or scared is tough. Grounding techniques, like noticing 5 things you can see around you, might help bring you back to the present.",
        "I hear that you're worried. Let's try to focus on what's in your control right now. Deep breathing can help calm your nervous system.",
        "It's scary to feel that way. Remember that anxiety often passes, even when it feels permanent. Drink a glass of cold water and take slow breaths."
    ],
    "surprise": [
        "It sounds like things caught you off guard. Take a moment to process what just happened.",
        "That must have been unexpected! Give yourself time to adjust to it.",
        "Life is full of unexpected moments. How are you feeling about it now that the initial shock is settling?"
    ]
}

# 4. Chat Interface Loop
print("\n=========================================")
print("=== Mental Health Support Assistant ===")
print("Type 'quit' or 'exit' to end the conversation.")
print("=========================================")

while True:
    user_input = input("You: ")
    if user_input.lower() in ['quit', 'exit']:
        print("\nBot: Take care! Remember you are not alone.")
        break
    if not user_input.strip(): continue
    
    # Get BERT Emotion
    bert_result = emotion_classifier(user_input)[0]
    emotion = bert_result['label']
    confidence = bert_result['score']
    
    print(f"\n[System Detected Emotion: {emotion.upper()} ({confidence*100:.1f}%)]")
    
    # Get Rule-Based Response
    if emotion in empathy_responses:
        response = random.choice(empathy_responses[emotion])
    else:
        response = "I hear you. Thank you for sharing that with me."
        
    print(f"Bot: {response}\n")
```
